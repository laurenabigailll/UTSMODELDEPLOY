```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
import pickle
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.histplot(data=df, x='CreditScore', kde=True, bins=30)
plt.title('Distribution of Credit Score')

plt.subplot(2, 2, 2)
sns.histplot(data=df, x='Age', kde=True, bins=30)
plt.title('Distribution of Age')

plt.subplot(2, 2, 3)
sns.histplot(data=df, x='Balance', kde=True, bins=30)
plt.title('Distribution of Balance')

plt.tight_layout()
plt.show()
```

```python
input = df.drop('churn', axis = 1)
output = df['churn']

x_train, x_test, y_train, y_test = train_test_split(input, output, test_size = 0.2, random_state = 42)
```

```python
print('x_train: ', x_train.shape)
print('y_train: ', y_train.shape)
print('x_test: ', x_test.shape)
print('y_test: ', y_test.shape)
```

```python
pilih = ['Gender', 'Geography']
OneHotEncoder = pd.get_dummies(x_train[pilih])
x_train = pd.concat([x_train.drop(pilih, axis=1), OneHotEncoder], axis=1)
```

```python
x_train.head()
```

```python
OneHotEncoder = pd.get_dummies(x_test[pilih])
x_test = pd.concat([x_test.drop(pilih, axis=1), OneHotEncoder], axis=1)
```

```python
x_test.head()
```

```python
x_test.head()
```

```python
use = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']
Scaler = StandardScaler()
x_train[use] = Scaler.fit_transform(x_train[use])
x_test[use] = Scaler.transform(x_test[use])
```

```python
rf_classifier = RandomForestClassifier(criterion = 'gini', max_depth = 4, random_state= 42)
rf_classifier.fit(x_train, y_train)
```

```python
y_predict = rf_classifier.predict(x_test)
print(classification_report(y_test, y_predict))
```

```python
param_grid = {'n_estimators': [50, 100],
              'criterion': ['gini', 'entropy'],
              'max_depth': [None, 10]}
```

```python
rf_classifier = RandomForestClassifier(random_state = 42)
rf_classifier = GridSearchCV(rf_classifier,
                                   param_grid = param_grid,
                                   scoring = 'accuracy',
                                   cv = 5)
```

```python
rf_classifier.fit(x_train,y_train)
print("Tuned Hyperparameters: ", rf_classifier.best_params_)
print("Accuracy: ", rf_classifier.best_score_)
```

```python
xgb_classifier = XGBClassifier(random_state=42)
xgb_classifier.fit(x_train, y_train)
```

```python
y_predict_xgb = xgb_classifier.predict(x_test)
print(classification_report(y_test, y_predict_xgb))
```

```python
param_grid_xgb = {'n_estimators': [50, 100],
                  'max_depth': [3, 5],
                  'learning_rate': [0.1, 0.01],
                  'subsample': [0.5, 0.7]
                  }
```

```python
xgb_classifier = XGBClassifier(random_state=42)
xgb_classifier = GridSearchCV(xgb_classifier,
                                    param_grid=param_grid_xgb,
                                    scoring='accuracy',
                                    cv=5)
```

```python
xgb_classifier.fit(x_train, y_train)
print("Tuned Hyperparameters: ", xgb_classifier.best_params_)
print("Accuracy: ", xgb_classifier.best_score_)
```

```python
rfclassifier = rf_classifier.best_estimator_
y_predict = rfclassifier.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
xgbclassifier = xgb_classifier.best_estimator_
y_predict_xgb_tuned = xgbclassifier.predict(x_test)
print(classification_report(y_test, y_predict_xgb_tuned))
```
